### 2.4 Finding Outliers in any of the numerical columns

Outliers are data points that differ significantly from the rest of the observations in a column. They appear as **extreme values** that are much **higher** or **lower** than the majority of the data

Finding Outliers using IQR method.

The **IQR (Interquartile Range)** method is a statistical technique used to identify outliers by measuring the spread of the middle 50% of data.

1. **Num_Vehicles_Involved** Column

```python
# Num_Vehicles_Involved Column

# first quartile(25th percentile)
Q1 = df['Num_Vehicles_Involved'].quantile(0.25)

# third quartile(75th percentile)
Q3 = df['Num_Vehicles_Involved'].quantile(0.75)

# IQR calculation
IQR = Q3 - Q1

# define lower bound
lower_bound = Q1 - 1.5 * IQR

# define upper bound
upper_bound = Q3 + 1.5 * IQR

lower_bound, upper_bound
```

```python
# filter rows where Num_Vehicles_Involved is outside bounds

outliers = df[(df['Num_Vehicles_Involved'] < lower_bound) | (df['Num_Vehicles_Involved'] > upper_bound)]

outliers
```

### 3.1.3 Multivariate Analysis
Analysis of three or more variables simultaneously to understand complex interactions and patterns.

1. Weather Conditions and Light Conditions and Num_Casualties

```python
weather_light_casualities = df.groupby(['Weather_Conditions', 'Light_Conditions']).agg(Total_Accidents  = ('Num_Casualties', 'count'),Avg_Casualties   = ('Num_Casualties', 'mean')).round(2)

print(weather_light_casualities)
```

The result shows **Rainy** weather causes the **most** accidents across all light
conditions **bold text** with **462** total incidents. The most dangerous combination is
**Rainy + Night** with an average of **1.55 casualties** per accident, while
**Foggy + Night** is comparatively safer with only **57 accidents**.

2. Type of Road and Road Conditions and Vehicle Speed

```python
road_type_condtion_speed = df.groupby(['Type_of_Road', 'Road_Conditions']).agg(Total_Accidents      = ('Vehicle_Speed', 'count'),Avg_Speed            = ('Vehicle_Speed', 'mean'),Avg_Speed_Difference = ('Speed_Difference', 'mean')).round(2)

print(road_type_condtion_speed)
```

**Rural** roads record **fewer** accidents **336** compared to **Highway** - **645**
